#### Joining

```python
left = pd.DataFrame({'key': range(5), 
                           'left_value': ['a', 'b', 'c', 'd', 'e']})
right = pd.DataFrame({'key': range(2, 7), 
                           'right_value': ['f', 'g', 'h', 'i', 'j']})
print(left)
print('\n')
print(right)
```

```python
# Inner join
pd.merge(left, right, on='key', how='inner')
```

```python
# Left join
pd.merge(left, right, on='key', how='left')
```

```python
# Left join
pd.merge(left, right, on='key', how='right')
```

```python
# Full outer
# Left join
pd.merge(left, right, on='key', how='outer')
```

#### Combining

```python
# pandas.concat takes a list of Series or DataFrames and returns a Series or DataFrame of the concatenated objects
pd.concat([left, right], sort=False)
```

```python
# By default, the function will vertically append the objects to one another, combining columns with the same name
# Objects can be concatentated side-by-side using the function's axis parameter
pd.concat([left, right], axis=1)
```

#### Group by

```python
# Get file
filename = '03-thanksgiving-2015-poll-data.csv'
data = pd.read_csv(filename)
data.head()
```

```python
# Use groupby function
by_main_dish = data.groupby('What is typically the main dish at your Thanksgiving dinner?')
by_main_dish
```

```python
# Calling coutn returns the total number of NOT NULL values within each column
by_main_dish.count()
```

```python
# Get distinct titles using .nunique() then .sort_values()
by_main_dish['How is the main dish typically cooked?'].nunique().sort_values(ascending=False)
```
